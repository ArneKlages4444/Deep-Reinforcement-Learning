```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas

name = 'setup'
```

```python
df_sac_her_sparse_1 = pandas.read_csv('Test_Dense_Sparse_SAC_HER/SAC+HER sparse reward_1.csv')
df_sac_her_sparse_2 = pandas.read_csv('Test_Dense_Sparse_SAC_HER/SAC+HER sparse reward_2.csv')
df_sac_her_sparse_3 = pandas.read_csv('Test_Dense_Sparse_SAC_HER/SAC+HER sparse reward_3.csv')
df_sac_her_sparse_4 = pandas.read_csv('Test_Dense_Sparse_SAC_HER/SAC+HER sparse reward_4.csv')
df_sac_her_sparse_5 = pandas.read_csv('Test_Dense_Sparse_SAC_HER/SAC+HER sparse reward_5.csv')
df_sac_her_sparse = df_sac_her_sparse_1.append(df_sac_her_sparse_2).append(df_sac_her_sparse_3).append(df_sac_her_sparse_4).append(df_sac_her_sparse_5)
df_sac_her_sparse[name] = ['SAC+HER sparse'] * len(df_sac_her_sparse)

df_sac_sparse_1 = pandas.read_csv('Test_Dense_Sparse_SAC_HER/SAC sparse reward_1.csv')
df_sac_sparse_2 = pandas.read_csv('Test_Dense_Sparse_SAC_HER/SAC sparse reward_2.csv')
df_sac_sparse_3 = pandas.read_csv('Test_Dense_Sparse_SAC_HER/SAC sparse reward_3.csv')
df_sac_sparse_4 = pandas.read_csv('Test_Dense_Sparse_SAC_HER/SAC sparse reward_4.csv')
df_sac_sparse_5 = pandas.read_csv('Test_Dense_Sparse_SAC_HER/SAC sparse reward_5.csv')
df_sac_sparse = df_sac_sparse_1.append(df_sac_sparse_2).append(df_sac_sparse_3).append(df_sac_sparse_4).append(df_sac_sparse_5)
df_sac_sparse[name] = ['SAC sparse'] * len(df_sac_sparse)

df_sac_dense_1 = pandas.read_csv('Test_Dense_Sparse_SAC_HER/SAC dense reward_1.csv')
df_sac_dense_2 = pandas.read_csv('Test_Dense_Sparse_SAC_HER/SAC dense reward_2.csv')
df_sac_dense_3 = pandas.read_csv('Test_Dense_Sparse_SAC_HER/SAC dense reward_3.csv')
df_sac_dense_4 = pandas.read_csv('Test_Dense_Sparse_SAC_HER/SAC dense reward_4.csv')
df_sac_dense_5 = pandas.read_csv('Test_Dense_Sparse_SAC_HER/SAC dense reward_5.csv')
df_sac_dense = df_sac_dense_1.append(df_sac_dense_2).append(df_sac_dense_3).append(df_sac_dense_4).append(df_sac_dense_5)
df_sac_dense[name] = ['SAC reward shaping'] * len(df_sac_dense)

ds_t_1 = df_sac_her_sparse.append(df_sac_sparse).append(df_sac_dense)
```

```python
sns.set_theme(style="darkgrid")
sns.lineplot(x="epoch", y="success rate", hue=name, data=ds_t_1)
plt.savefig('sac_her_dense_sparse_success.pdf')
```

```python
df_final_1 = pandas.read_csv('Test_Final_K_Final_Future/SAC+HER final_1.csv')
df_final_2 = pandas.read_csv('Test_Final_K_Final_Future/SAC+HER final_2.csv')
df_final_3 = pandas.read_csv('Test_Final_K_Final_Future/SAC+HER final_3.csv')
df_final_4 = pandas.read_csv('Test_Final_K_Final_Future/SAC+HER final_4.csv')
df_final_5 = pandas.read_csv('Test_Final_K_Final_Future/SAC+HER final_5.csv')
df_final = df_final_1.append(df_final_2).append(df_final_3).append(df_final_4).append(df_final_5)
df_final[name] = ['SAC+HER final'] * len(df_final)

df_k_final_1 = pandas.read_csv('Test_Final_K_Final_Future/SAC+HER k final_1.csv')
df_k_final_2 = pandas.read_csv('Test_Final_K_Final_Future/SAC+HER k final_2.csv')
df_k_final_3 = pandas.read_csv('Test_Final_K_Final_Future/SAC+HER k final_3.csv')
df_k_final_4 = pandas.read_csv('Test_Final_K_Final_Future/SAC+HER k final_4.csv')
df_k_final_5 = pandas.read_csv('Test_Final_K_Final_Future/SAC+HER k final_5.csv')
df_k_final = df_k_final_1.append(df_k_final_2).append(df_k_final_3).append(df_k_final_4).append(df_k_final_5)
df_k_final[name] = ['SAC+HER k final'] * len(df_k_final)

df_future_1 = pandas.read_csv('Test_Final_K_Final_Future/SAC+HER future_1.csv')
df_future_2 = pandas.read_csv('Test_Final_K_Final_Future/SAC+HER future_2.csv')
df_future_3 = pandas.read_csv('Test_Final_K_Final_Future/SAC+HER future_3.csv')
df_future_4 = pandas.read_csv('Test_Final_K_Final_Future/SAC+HER future_4.csv')
df_future_5 = pandas.read_csv('Test_Final_K_Final_Future/SAC+HER future_5.csv')
df_future = df_future_1.append(df_future_2).append(df_future_3).append(df_future_4).append(df_future_5)
df_future[name] = ['SAC+HER future'] * len(df_future)

ds_t_2 = df_final.append(df_k_final).append(df_future)
```

```python
sns.set_theme(style="darkgrid")
sns.lineplot(x="epoch", y="success rate", hue=name, data=ds_t_2)
plt.savefig('final_k_final_future_success.pdf')
```
